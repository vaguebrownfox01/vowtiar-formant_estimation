```python
import pandas as pd
import numpy as np 


from sklearn import metrics
```

```python
synth_vowel_csv = "exports/e_synth-vowels_formant_estimation_vowlim100.csv"
synth_vowel_filt_csv = "exports/e_synth-vowels_formant_estimation_vowlim100_filt.csv"
synth_df = pd.read_csv(synth_vowel_csv)
filt_df = pd.read_csv(synth_vowel_filt_csv)
```

```python
cols_to_use = filt_df.columns.difference(synth_df.columns)
```

```python
filt_df.columns
```

```python
temp = pd.merge(synth_df, filt_df[cols_to_use], left_index=True, right_index=True, how='outer')
temp.columns
```

```python
df = temp.dropna()
df
```

```python
metrics_dict = {
    "vowel_name": [],
    "sex": [],
    "F1_praat_rms": [],
    "F2_praat_rms": [],
    "F3_praat_rms": [],
    "F1_praat_filter_rms": [],
    "F2_praat_filter_rms": [],
    "F3_praat_filter_rms": [],

}

for vowel in pd.unique(df["vowel_name"]):
    for gender in  pd.unique(df["sex"]):
        f1_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_mean_praat_base"]
        f2_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_mean_praat_base"]
        f3_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_mean_praat_base"]

        f1_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_mean_synthvow_praat"]
        f2_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_mean_synthvow_praat"]
        f3_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_mean_synthvow_praat"]

        f1_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_mean_synthvow_filter_praat"]
        f2_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_mean_synthvow_filter_praat"]
        f3_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_mean_synthvow_filter_praat"]

        metrics_dict["vowel_name"].append(vowel)
        metrics_dict["sex"].append(gender)
        metrics_dict["F1_praat_rms"].append(metrics.mean_squared_error(f1_gt, f1_praat, squared=False))
        metrics_dict["F2_praat_rms"].append(metrics.mean_squared_error(f2_gt, f2_praat, squared=False))
        metrics_dict["F3_praat_rms"].append(metrics.mean_squared_error(f3_gt, f3_praat, squared=False))

        metrics_dict["F1_praat_filter_rms"].append(metrics.mean_squared_error(f1_gt, f1_deep, squared=False))
        metrics_dict["F2_praat_filter_rms"].append(metrics.mean_squared_error(f2_gt, f2_deep, squared=False))
        metrics_dict["F3_praat_filter_rms"].append(metrics.mean_squared_error(f3_gt, f3_deep, squared=False))


df_metrics = pd.DataFrame(metrics_dict)
df_metrics
```

```python
df_metrics.to_csv("./exports/g_filter_error_metrics.csv")
```

```python
df_metrics_trans = df_metrics.T
df_metrics_trans = df_metrics_trans.iloc[2:,:]
df_metrics_trans.set_axis(df_metrics["vowel_name"] + df_metrics["sex"], axis=1, inplace=True)
```

```python
df_metrics_trans.to_csv("./exports/g_filter_error_metrics_trans.csv")
```
