```python
from scipy import stats
import pandas as pd
import numpy as np


```

```python
df_spec = df[(df["vowel_name"] == "iy") & (df["sex"] == "F")]
stats.ttest_ind(df_spec["F1_praat_error"],df_spec["F1_praat_error"])
```

```python
res = []
for vowel in pd.unique(df["vowel_name"]):
    for gender in  pd.unique(df["sex"]):
        df_spec = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]
        for i in range(1,4):
            wil_test = stats.wilcoxon(df_spec[f"F{i}_praat_error"],df_spec[f"F{i}_czt-lpc_error"])[-1]
            t_test = stats.ttest_ind(df_spec[f"F{i}_praat_error"],df_spec[f"F{i}_czt-lpc_error"])[-1]
            # if wil_test >= 0.05 or t_test >= 0.05:

            re = { "Vowel": vowel ,"Gender": gender , "Formant" : f"F{i}"  ,"wil_test": round(wil_test, 5) ,"t_test": round(t_test, 5) }

            res.append(re)

res_df = pd.DataFrame(res)
filt = res_df[(res_df["wil_test"] >= 0.05) | (res_df["t_test"] >= 0.05)]
filt

```
