```python
import scipy.io
import pandas as pd
from sklearn import metrics
import numpy as np
```

```python
mat_export_path = "/home/jeevan/Jeevan_K/Projects/Asquire/Vowtiar-Quest/vowtiar-formant_estimation/matlab/formant_estimation_lpc_1secwin.mat"
synth_vowel_formant_csv_path = "/home/jeevan/Jeevan_K/Projects/Asquire/Vowtiar-Quest/vowtiar-formant_estimation/exports/e_synth-vowels_formant_estimation_vowlim100_fullwin.csv"
```

```python
synth_vowel_df = pd.read_csv(synth_vowel_formant_csv_path); synth_vowel_df
```

```python
mat = scipy.io.loadmat(mat_export_path); mat
```

```python
mat_df = pd.DataFrame({
    "index": mat["index"].flatten(),
    "F1_truth": mat["formants_truth"][:,0],
    "F2_truth": mat["formants_truth"][:,1],
    "F1_truth": mat["formants_truth"][:,2],
    "F1_lpc": mat["formants_pred"][:,0],
    "F2_lpc": mat["formants_pred"][:,1],
    "F3_lpc": mat["formants_pred"][:,2],
    "F1_error": mat["res_errors"][:,0],
    "F2_error": mat["res_errors"][:,1],
    "F3_error": mat["res_errors"][:,2],
})
```

```python
df = pd.merge(synth_vowel_df, mat_df, on="index")
df.columns
```

```python
for i in range(1,4):
    df[f"F{i}_praat_error"] = np.abs(df[f"F{i}_mean_praat_base"] - df[f"F{i}_mean_synthvow_filter_praat"])
    df[f"F{i}_czt-lpc_error"] = np.abs(df[f"F{i}_mean_praat_base"] - df[f"F{i}_lpc"])
df
```

```python
metrics_dict = {
    "vowel_name": [],
    "sex": [],
    "F1_praat_error_mean": [],
    "F2_praat_error_mean": [],
    "F3_praat_error_mean": [],
    "F1_czt_lpc_error_mean": [],
    "F2_czt_lpc_error_mean": [],
    "F3_czt_lpc_error_mean": [],
    "F1_praat_error_std": [],
    "F2_praat_error_std": [],
    "F3_praat_error_std": [],
    "F1_czt_lpc_error_std": [],
    "F2_czt_lpc_error_std": [],
    "F3_czt_lpc_error_std": [],


}

for vowel in pd.unique(df["vowel_name"]):
    for gender in  pd.unique(df["sex"]):
        # f1_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_mean_praat_base"]
        # f2_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_mean_praat_base"]
        # f3_gt = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_mean_praat_base"]

        # f1_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_mean_synthvow_praat"]
        # f2_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_mean_synthvow_praat"]
        # f3_praat = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_mean_synthvow_praat"]

        # f1_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F1_lpc"]
        # f2_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F2_lpc"]
        # f3_deep = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]["F3_lpc"]

        df_spec = df[(df["vowel_name"] == vowel) & (df["sex"] == gender)]
        metrics_dict["vowel_name"].append(vowel)
        metrics_dict["sex"].append(gender)
        for i in range(1,4):
            metrics_dict[f"F{i}_praat_error_mean"].append(np.mean(df_spec[f"F{i}_praat_error"]))
            metrics_dict[f"F{i}_czt_lpc_error_mean"].append(np.mean(df_spec[f"F{i}_czt-lpc_error"]))
            metrics_dict[f"F{i}_praat_error_std"].append(np.std(df_spec[f"F{i}_praat_error"]))
            metrics_dict[f"F{i}_czt_lpc_error_std"].append(np.std(df_spec[f"F{i}_czt-lpc_error"]))
        

df_metrics = pd.DataFrame(metrics_dict)
df_metrics
```

```python
df_metrics_trans = df_metrics.T
df_metrics_trans = df_metrics_trans.iloc[2:,:]
df_metrics_trans.set_axis(df_metrics["vowel_name"] + "_" + df_metrics["sex"], axis=1, inplace=True)

```

```python
df_metrics_trans.index.names = ["metrics"]
df_metrics_trans
```

```python
df_metrics_trans.to_csv("./exports/h_filter_lpc_error_metrics_1secwin.csv", index=True)
```
